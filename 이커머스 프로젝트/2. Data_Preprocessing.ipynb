### 결측값 제거

크롤링 코드 작성 시 크롤링하지 못한 정보는 '없음'으로 저장하게 했기 때문에 '없음'의 갯수로 결측값의 갯수를 파악합니다.

```python
# 데이터 컬럼별 없음(결측치) 개수 파악

col_name= df.columns    # 컬럼명 저장
from collections import Counter

for i in col_name:
    print(i,Counter(df[i])["없음"])
    
```

```python
### 별점, 피부정보, 피부타입, 피부고민, 자극도값이 없음은 제거한 데이터를 df2에 저장

df2 = df[(df['별점'] != '없음') &
   (df['피부정보'] != '없음') &
   (df['피부타입'] != '없음') &
   (df['피부고민'] != '없음') &
   (df['자극도'] != '없음')]
df2.reset_index(drop=True, inplace=True)
df2.shape
```

```python
# 데이터 컬럼별 없음(결측치) 개수 파악

col_name= df2.columns    #컬럼명 저장
from collections import Counter

for i in col_name:
    print(i,Counter(df2[i])["없음"])
```

### 정가, 할인가 데이터 타입 변경

정가,할인가의 쉼표를 제거하고 데이터 타입을 int64로 바꿔줍니다.

```python
df2.head()
```

```python
def extract_number(x):
    try:
        x = x.replace(',', '')
    except:
        pass
    x = int(x)
    return x

df2['정가'] = df2['정가'].apply(extract_number)
df2['할인가'] = df2['할인가'].apply(extract_number)
```

```python
df2.head()
```

```python
df2.dtypes
```

### 아이디 컬럼 삭제

모델 학습 시 아이디 정보는 필요하지 않기 때문에 제거해줍니다.

```python
df2 = df2.drop('아이디', axis=1) #'아이디'에 일치하는 열(axis=1)의 데이터를 제거
df2.head()
```

### 별점에서 숫자만 남기기

현재 별점은 5점만점에 X점과 같은 형식으로 되어있습니다. 여기서 X만 남기고 데이터 타입을 int64로 바꿔줍니다.

```python
import re # 정규 표현식을 사용

def extract_rate(x):
    rate = re.findall('\d', x)[1]
    return int(rate)
    

df2['별점'] = df2['별점'].apply(extract_rate)
df2.head()
```

```python
df2.dtypes
```

### 피부정보 컬럼을 여러컬럼으로 구분

피부정보에는 최대 피부타입, 피부톤, 2개의 피부고민 총 4개로 이루어져 있습니다.  
해당 텍스트는 공백으로 구분되어있고 이를 이용해 4개의 컬럼(user피부타입/user피부톤/user피부고민1/user피부고민2)으로 만들어 줍니다.

```python
split_columns = df2["피부정보"].str.split(expand=True)
split_columns.columns = ["user피부타입", "user피부톤", "user피부고민1", "user피부고민2"]

df2["user피부타입"] = split_columns["user피부타입"]
df2["user피부톤"] = split_columns["user피부톤"]
df2["user피부고민1"] = split_columns["user피부고민1"]
df2["user피부고민2"] = split_columns["user피부고민2"] 

df2.head()
```

#### user피부타입 컬럼에서 복합성/건성/지성/민감성/트러블성/약건성/중성 을 제외하고 모두 np.NaN 처리 해줍니다.  

```python
#user피부타입에서 해당하지 않는 값은 nan 처리
def to_nan(x):
    if x not in ['복합성', '건성', '지성', '민감성', '트러블성', '약건성', '중성']:
        return np.NaN
    else:
        return x
        
        
df2["user피부타입"] = df2["user피부타입"].apply(to_nan)
df2["user피부타입"].value_counts() # value_counts() : 해당 열의 고유값과 빈도를 계산하는 메소드
```

#### user피부고민2의 결측값을 user피부톤 컬럼값으로 채운 후  (웜톤/쿨톤/여름쿨톤/봄웜톤/가을웜톤/겨울쿨톤) 값을  np.NaN 처리 해줍니다.

```python
df2["user피부고민2"].value_counts()
```

```python
df2["user피부고민2"]= df2["user피부고민2"].fillna(df2["user피부톤"])
df2["user피부고민2"].value_counts()
```

```python
def to_nan_피부고민2(x):
    if x in ["웜톤", "쿨톤", "여름쿨톤", "봄웜톤", "가을웜톤", "겨울쿨톤"]:
        return np.NaN
    else:
        return x
        
        
df2["user피부고민2"] = df2["user피부고민2"].apply(to_nan_피부고민2)
df2["user피부고민2"].value_counts()
```

### user피부타입, usesr피부고민1, user피부고민2 에서 결측값 제거

```python
df2 = df2.dropna(subset=["user피부타입", "user피부고민1", "user피부고민2"])
df2.isnull().sum()
```

### df2를 df3에 복사한 후 상품명컬럼의 값을 간단하게 변경해줍니다.
<pre>
[8월 올영픽/1+1기획] 비원츠 피토 콜라겐 아이 세럼 스틱 15ml 1+1 기획                         : bewants
[홍현희 PICK/ 한정기획] 구달 청귤 비타C 잡티케어 세럼 50ml+50ml 리필기획                     : goodal
[뽐니PICK/한정기획] 토리든 다이브인 세럼 50ml 리필 한정 기획(+50ml 리필팩+수딩 크림 20ml)     : torriden
[8월 올영픽/1+1단독] 성분에디터 그린토마토 포어 리프팅 앰플 플러스 30ml 더블 기획             : editor
[8월 올영픽/대용량한정기획]메이크프렘 인테카 수딩크림 100ml 기획(+크림50ml+크림31ml)          : prem
</pre>

```python
df3=df2.copy()

#상품명 간소화
def name_min(x):
    if '비원츠' in x:
        return "bewants"
    elif '구달' in x:
        return "goodal"
    elif '토리든' in x:
        return "torriden"
    elif '성분에디터' in x:
        return "editor"
    elif '메이크프렘' in x:
        return "prem"
    else:
        return np.NaN

df3["상품명"] = df3["상품명"].apply(name_min)
df3.head()
```

```python
df3["상품명"].value_counts() # 상품명에 지정한 명칭이 들어갔는지 검사
```

```python
df3["상품명"].isnull().sum() # 상품명에 오류가 난 행이 있는지 검사
```

### 직접 함수를 만들어 자극도 컬럼을 레이블인코딩 합니다.
* 자극없이 순해요 -> 0
* 보통이에요 -> 1
* 자극이 느껴져요 -> 2

```python
def 자극도인코딩(x):
    if x == '자극없이 순해요':
        return 0
    elif x == '보통이에요':
        return 1
    elif x == '자극이 느껴져요':
        return 2
    else:
        return np.NaN

df3["자극도"] = df3["자극도"].apply(자극도인코딩)
df3["자극도"].value_counts()
```

```python
df3["자극도"].isnull().sum() # 자극도에 잘못된 값이 있었는지 체크
```
